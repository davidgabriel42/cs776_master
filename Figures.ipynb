```python
scale_times = [2.57,2.57,2.58,1.83,1.56,1.37,1.65,1.38,1.38,1.38,1.37]
six_nodes = [1.23,1.3,1.29,1.22,1.29,1.29,1.29,1.29,1.3,1.3]
four_gpus = [2.4,2.43,2.46,2.48,2.59,2.56,2.6,2.65,2.66,2.85,2.75]
cpu_only = [19.17,19.3,19.35,19.54,19.48,19.54,19.41,19.48,19.43,19.23]
best_ever = min(six_nodes)* np.ones(len(scale_times))
print(len(scale_times))
print(len(range(0,11)))
```
